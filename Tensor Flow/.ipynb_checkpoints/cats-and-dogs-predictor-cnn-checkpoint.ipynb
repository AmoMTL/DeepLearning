```python
training_data = []

datadir = r'C:\Users\Amo\Desktop\Amogha\Python\DeepLearning\Tensor Flow\Datasets\PetImages'
categories= ['Dog','Cat']

img_size = 50 

# O will represent "Dog" and 1 will represent "Cat"

def create_training_data():
    for category in categories:
        path = os.path.join(datadir,category)
        pet_class = categories.index(category)
        for img in os.listdir(path):
            try:
                img_array = cv2.imread(os.path.join(path,img),cv2.IMREAD_GRAYSCALE)
                resized_img_array = cv2.resize(img_array,(img_size,img_size))
                training_data.append([resized_img_array,pet_class])
            except Exception as e:
                pass
```

```python
create_training_data()
```

```python
print(len(training_data))
```

```python
# Shuffle the loaded data
random.shuffle(training_data)
```

```python
# Constructing the training variables

X = []
y = []

for img, label in training_data:
    X.append(img)
    y.append(label)

X = np.array(X).reshape(-1,img_size,img_size,1) 
```

```python
# Save the preprocessed dataset
import pickle

pickle_out = open("cats_dogs_X.pickle","wb")
pickle.dump(X,pickle_out)
pickle_out.close()

pickle_out = open("cats_dogs_y.pickle","wb")
pickle.dump(y,pickle_out)
pickle_out.close()
```
